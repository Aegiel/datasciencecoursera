# Code
Below you can find the code that I used as well as its output:

```python
import pandas as pd
import numpy as np
import seaborn
import scipy
import matplotlib.pyplot as plt

data=pd.read_csv("nesarc.csv",low_memory=False)

data["AGE"] = pd.to_numeric(data["AGE"], errors="coerce")
data["S1Q24LB"] = pd.to_numeric(data["S1Q24LB"], errors="coerce")

data['AGE']=data['AGE'].replace(' ', np.nan)
data['S1Q24LB']=data['S1Q24LB'].replace(999, np.nan)

data_clean = data.dropna()

print ('association between age and weight')
print (scipy.stats.pearsonr(data_clean['AGE'], data_clean['S1Q24LB']))

scat1 = seaborn.regplot(x="AGE", y="S1Q24LB", fit_reg=True, data=data)
plt.xlabel('Age')
plt.ylabel('Weight (Lb)')
plt.title('Scatterplot for the Association Between Age and Weight')
```

# Examining the results

We started by selecting **Two Quantitative Variables** to perform our Pearson Correlation, which were *Age* and *Weight*.

The correlation coefficient obtained tells us there is a small negative correlation between the two variables (**-0.0242**), since it is rather close to zero, meaning that as x increases, y decreases. However we can see that it is significative because the P-value obtained is **7.927e-07**, which is smaller than 0.05.

Finally, when we square the correlation coefficient we get our Coefficient of Determination, which is interpreted as the proportion of the variance of the dependent variable that can be predicted by the independent one. In our case, (-0.0242)^2=**0.000584**. This means that 0.05% of the variance in the Weight is explained by the Age in the linear model.
